```python
import random
import string

vehicleCount = 50

def vhl_generator(size=2, chars= string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
from collections import defaultdict
dic = defaultdict(str)

for i in range(vehicleCount):
    vhlId = vhl_generator()
    if vhlId in dic:
        continue
    dic[vhlId] = '1'

dic
        
```
